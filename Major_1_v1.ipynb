## Background
----------------
At the quantum level, all particles of matter display wave-like properties. The wavelength of a particle, called the de Broglie wavelength, changes according to the particle’s momentum. Parameters for this calculation are the velocity of the particle v, the mass of the particle at rest m0, the speed of light
$c = 3 × 10^8 $m/s, and Planck’s constant h = 6.63 × 10−34 J ·s. The
formula for calculating the de Broglie wavelength (in metres) is:
$$\lambda = \frac{h\sqrt{1 - \frac{v^2}{c^2}}}{m_0v}$$
Sample values are as follows: $v = 2.2 \times 10^6$ m/s and $m_0 = 9.11 \times 10^{-31}$ kg.

## Calculate A - de Broglie wavelength using $v$ and $m_0$ (4 Points)

```python
def A(v, m0):
    """This function computes the de Broglie wavelength using v and m0"""
    ### BEGIN SOLUTION
    import math
    h = 6.63*10**-34
    c = 3*10**8
    return (h*math.sqrt(1-(v**2/c**2)))/(m0 * v)
    ### END SOLUTION
```

```python
"""Check that A returns the correct output for several inputs"""
assert A(2.2*10**6, 9.11*10**-31) == 3.3079641374462706e-10
assert A(2.8e8,3.45e-34) == 2.464019508026347e-09
```

## Calculate B - de Broglie wavelength using $v$ and $5m_0$ (4 Points)

```python
def B(v, m0):
    """This function computes the de Broglie wavelength using v and 5m0"""
    ### BEGIN SOLUTION
    import math
    h = 6.63*10**-34
    c = 3*10**8
    return (h*math.sqrt(1-(v**2/c**2)))/((5*m0) * v)
    ### END SOLUTION
```

```python
assert B(2.2*10**6, 9.11*10**-31) == 6.61592827489254e-11
assert B(2.8e8,3.45e-34) == 4.928039016052694e-10
```

## Calculate C - Relative difference between A and B (4 Points)

```python
def C(v, m0):
    """This function computes the relative difference between A and B"""
    ### BEGIN SOLUTION
    return 100.0 * abs((A(v, m0) - B(v, m0)) / B(v, m0))
    ### END SOLUTION
```

```python
assert C(2.2*10**6, 9.11*10**-31) == 400.0000000000001
assert C(2.8e8,3.45e-34) == 400.0
```
