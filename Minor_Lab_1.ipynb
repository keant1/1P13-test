# e_convergence Function - 1 term at a time (5 Points)

--------------
REMINDER: You can run a code cell using the run cell button on the toolbar above. This will allow you to test your code as you complete it. HINT: The `input` statement should go on the line before the function definition.

````python
def e_convergence(n):
    """This function illustrates the convergence of (1 + 1/n)^n, and returns the nth term of convergence"""
    ### BEGIN SOLUTION
    return (1 + (1.0/n)) ** n
    ### END SOLUTION
    
````

````python
"""Check that e_convergence returns the correct output for several inputs"""
assert e_convergence(1) == 2.0
assert e_convergence(2) == 2.25
assert e_convergence(3) == 2.37037037037037
assert e_convergence(4) == 2.44140625
assert e_convergence(5) == 2.4883199999999994
````

# e_convergence_list Function - All terms from 1 to n (5 Points)
-------------

````python
def e_convergence_list(n):
    """This function illustrates the convergence of (1 + 1/n)^n, and returns a list of terms 1 to n"""
    ### BEGIN SOLUTION
    value_list = []
    for n in range(1, n+1):
        value_list.append((1 + (1.0/n)) ** n)
    return value_list
    ### END SOLUTION
````

````python
assert e_convergence_list(1) == [2.0]
assert e_convergence_list(2) == [2.0, 2.25]
assert e_convergence_list(3) == [2.0, 2.25, 2.37037037037037]
assert e_convergence_list(4) == [2.0, 2.25, 2.37037037037037, 2.44140625]
assert e_convergence_list(5) == [2.0, 2.25, 2.37037037037037, 2.44140625, 2.4883199999999994]
````
